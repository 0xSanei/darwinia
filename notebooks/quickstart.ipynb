## 4. Inspect Champion DNA

```python
import json, glob

champions = sorted(glob.glob('output/champions/*.json'))
if champions:
    with open(champions[-1]) as f:
        champ = json.load(f)
    print(f"Champion ID: {champ['id']}")
    print(f"Fitness: {champ['fitness']:.4f}")
    print(f"Generation: {champ['generation']}")
    print(f"\nGenes:")
    for gene, val in champ['genes'].items():
        bar = '█' * int(val * 20) + '░' * (20 - int(val * 20))
        print(f"  {gene:25s} {bar} {val:.3f}")
else:
    print('No champions found. Run evolution first.')
```

## 6. Compare Archetypes Against Attacks

```python
from darwinia.core.dna import AgentDNA
from darwinia.arena.arena import AdversarialArena

archetypes = {
    'Trend Follower': AgentDNA.seed_trend_follower(),
    'Mean Reverter': AgentDNA.seed_mean_reverter(),
    'Conservative': AgentDNA.seed_conservative(),
    'Aggressive': AgentDNA.seed_aggressive(),
    'Random': AgentDNA.random(),
}

for name, dna in archetypes.items():
    arena = AdversarialArena({'rounds_per_test': 5})
    survival = arena.test_agent(dna, normal_data=None)
    attacks = {r.trap_type: '✅' if r.survived else '❌' for r in arena.history}
    print(f"{name:20s} | Survival: {survival:.0%} | {' '.join(f'{k}:{v}' for k,v in attacks.items())}")
```
